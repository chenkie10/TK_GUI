```python
import tkinter as tk
from tkinter import ttk

def on_combobox_change(event):
    # Function to be called when the combobox value changes
    selected_value = combobox.get()
    label.config(text=f"Selected value: {selected_value}")

# Create the main Tkinter window
root = tk.Tk()
root.title("Selection Box with Combobox")

# Create a list of options for the combobox
options = ["Option 1", "Option 2", "Option 3", "Option 4"]

# Create a Combobox widget
combobox = ttk.Combobox(root, values=options)
combobox.pack(pady=10)
combobox.set(options[0])  # Set the default selected option

# Bind the on_combobox_change function to the <<ComboboxSelected>> event
combobox.bind("<<ComboboxSelected>>", on_combobox_change)

# Create a label to display the selected value
label = tk.Label(root, text=f"Selected value: {options[0]}")
label.pack(pady=10)

# Run the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk

def button_clicked():
    print("Button Clicked!")

def create_custom_button(canvas, shape, coords, command):
    # Create a custom-shaped button on the canvas
    button = canvas.create_polygon(coords, outline='black', fill='lightblue', width=2)
    # Bind the button click event to the specified command
    canvas.tag_bind(button, '<Button-1>', lambda event: command())

# Create the main Tkinter window
root = tk.Tk()
root.title("Custom Shaped Button")

# Create a Canvas widget
canvas = tk.Canvas(root, width=300, height=200)
canvas.pack()

# Define the coordinates for a triangle (example shape)
triangle_coords = [50, 150, 150, 50, 250, 150]

# Create a custom-shaped button (triangle) on the canvas
create_custom_button(canvas, 'triangle', triangle_coords, button_clicked)

# Run the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk

def on_resize(event):
    # Function to be called when the window is resized
    new_width = event.width
    new_height = event.height

    # Calculate new position for the button based on the window size
    button_x = new_width // 2
    button_y = new_height // 2

    # Set the new position of the button
    button.place(x=button_x, y=button_y)
    button_1.place(x=button_x, y=button_y)

# Create the main Tkinter window
root = tk.Tk()
root.title("Resizable Window")

# Create a button
button = tk.Button(root, text="Click me!")
button_1=tk.Button(root,text="new")

# Bind the on_resize function to the window resize event
root.bind("<Configure>", on_resize)

# Initially place the button at the center of the window
button.place(relx=0.5, rely=0.5, anchor="center")
button_1.place(x=100,y=200)

# Run the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk

def toggle_checkbox():
    # Toggle the state of the checkbox
    checkbox_state.set(not checkbox_state.get())

# Create the main Tkinter window
root = tk.Tk()
root.title("Colored Checkbox Example")

# Variable to store the state of the checkbox (0 for unchecked, 1 for checked)
checkbox_state = tk.IntVar()

# Create a Checkbutton with custom text color and background color
checkbox = tk.Checkbutton(root, text="Check me", variable=checkbox_state, command=toggle_checkbox,
                          fg="blue",  # Text color
                          bg="lightgray"  # Background color
                          )
checkbox.pack(pady=10)

# Run the Tkinter event loop
root.mainloop()
```
